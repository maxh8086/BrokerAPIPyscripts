```python
def getprvDayCandle(symbol):
    try:
        start_date = datetime.today() - timedelta(days=10)
        dayList = api.get_daily_price_series( 'NSE', symbol, startdate=start_date.timestamp(), enddate=None)
        newDayList= []
        for i in dayList:
            newDayList.append(json.loads(i))
        dayDf = pd.DataFrame(newDayList)
        dayDf['time'] = pd.to_datetime(dayDf['time']).apply(lambda x: x.date())   
        dayDf.sort_values(by = 'time',inplace=True)
        if  dayDf.iloc[-1].time == datetime.now().date():
            return dayDf.iloc[-2].to_dict()
        else:
            return dayDf.iloc[-1].to_dict()  
    except Exception as e:
        print(symbol, e)
```

```python
symbolCloseList = []
for i in filterSymbols.index:
    symbolRow = filterSymbols.loc[i]
    cnadleInfo = getprvDayCandle(symbolRow.TradingSymbol)
    cnadleInfo.update({'symbol' : symbolRow.Symbol ,'token' : symbolRow.Token  })
    symbolCloseList.append(cnadleInfo)
    
symbolCloseList
```
